#### Covariance Formula

$cov_{x,y}=\frac{\sum_{i=1}^{N}(x_{i}-\bar{x})(y_{i}-\bar{y})}{N-1}$

```python
cov_num_var = df[num_list].cov()
cov_num_var
```

#### Observations:

The covariance table helps us understand that all the numeric fields have a poistive correlation

The diagonal numbers represent the variance of each of the numeric fields

From this data it is difficult to understand and compare the correlation between the fields. Hence, we arrive at coefficient of correlation from this data.

#### Correlation Coefficient

 (X,Y) = cov (X,Y) / X.Y

```python
corr_num_var = df[num_list].corr()
corr_num_var
```

#### Observations:

Coefficient of correlation gives us a value between -1 to +1. If the value is more near to +1 or -1 it represents a string correlation. A value near to 0 mean weak correlation between the numeric fields.

```python
sns.heatmap(corr_num_var, annot=True)
```

#### Observations:

From above we see strong positive correlation between Runs_Scored and Mins_Played, Runs_Scored and Balls_Faced. We also observe a positive relation between Mins_Played and 4s, Balls_Faced and 4s
